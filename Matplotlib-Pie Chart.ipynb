```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
course = ["datascience","devops","python","AI","ml"]
stds = [50,20,30,40,10]
```

```python
plt.pie([10])
```

```python
plt.pie([10,30,45,33,25])
plt.show()
```

```python
plt.pie(stds)
plt.show()
```

```python
plt.pie(stds,labels = course)
plt.show()
```

```python
a = [0.1,0.05,0.05,0.05,0.05]
plt.pie(stds,labels = course,explode = a)

plt.show()
```

```python
a = [0.05,0.05,0.05,0.05,0.05]
col = ["r","yellow","g","b","y"]
plt.pie(stds,labels = course,explode = a,colors = col)

plt.show()
```

```python
a = [0.05,0.05,0.05,0.05,0.05]
col = ["r","yellow","g","b","y"]
plt.pie(stds,labels = course,explode = a,colors = col,autopct = "%0.1f%%")

plt.show()
```

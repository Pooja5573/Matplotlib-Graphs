```python
df2 = pd.read_csv("C:\\Users\\Satish\\Downloads\\DataSets\\weight_height.csv")
df2.head()
```

```python
x1 = df2["Height"][:500]
y1 = df2["Weight"][:500]
```

```python
plt.scatter(x1,y1)
```

```python
plt.figure(figsize = (15,8))
plt.scatter(x1,y1,color = "r",marker = "*", s = 100)
plt.xlabel ("Height")
plt.ylabel("Weight")
```

```python
plt.figure(figsize = (15,8))
plt.scatter(x1,y1,color = "r",marker = "*", s = 100,alpha = 0.2)
plt.xlabel ("Height")
plt.ylabel("Weight")
```

```python
x2 = df2["Height"][100:200]
y2 = df2["Weight"][100:200]
```

```python
plt.figure(figsize = (15,8))
plt.scatter(x1,y1,color = "r",marker = "*", s = 100,alpha = 0.2)
plt.scatter(x2,y2,color = "b",marker = "o", s = 100)

plt.xlabel ("Height")
plt.ylabel("Weight")
```

```python
plt.figure(figsize = (15,8))
plt.scatter(x2,y2,color = "yellow",marker = "*", s = 500,linewidth = 2,edgecolor = "r")
plt.xlabel ("Height")
plt.ylabel("Weight")
plt.legend(["Height vs Weight"])
```
